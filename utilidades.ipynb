```python
import pandas as pd
import json

f= open('data/grafo_autores.json') 
data_autores = json.load(f)
f.close()

salida_authorrank = pd.read_excel("data/salida_AuthorRank.xlsx",index_col= 0)

rankings = pd.read_excel("data/centralidades.xlsx",index_col= 0)
```

```python
salida_authorrank['ins'] = salida_authorrank['ins'].astype(str)
salida_authorrank

rankings.reset_index(inplace=True)
rankings.rename(columns={'index': 'ins'}, inplace=True)
rankings['ins'] = rankings['ins'].astype(str)
rankings['nombre'] = rankings['ins'].apply(lambda x: data_autores[x]["nombre_autor"] if x in data_autores else "Unknown")

centralidad_final = pd.merge(salida_authorrank, rankings, on='ins', how='inner', suffixes=('_salida_authorrank', '_rankings'))

centralidad_final.rename(columns={
    'ins': 'Nodo',
    'values': 'AuthorRank',
    'nombre_salida_authorrank': 'Nombre'
}, inplace=True)

# Eliminar la columna 'nombre_rankings'
centralidad_final.drop(columns=['nombre_rankings'], inplace=True)

# Reordenar las columnas
centralidad_final = centralidad_final[['Nodo', 'Nombre', 'AuthorRank', 'PageRank', 'Betweenness', 'Closeness']]

centralidad_final

```

```python
centralidad_final.to_excel('data/centralidad_final.xlsx')
```
